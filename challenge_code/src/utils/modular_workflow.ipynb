# Challenge Data : Leukemia Risk Prediction - Version Modulaire

Ce notebook utilise la nouvelle architecture modulaire du projet.

```python
# Configuration et imports
import sys
import os
sys.path.append('../src')

# Imports des modules personnalisés
from config import *
from data.load import load_all_data
from data.prepare import clean_target_data, prepare_enriched_dataset, prepare_features_and_target
from modeling.train import train_and_save_all_models
from modeling.evaluate import compare_models
from utils.helpers import set_seed, print_dataset_info
from visualization.plots import plot_correlation_matrix, create_visualization_report
```

```python
# Configuration initiale
set_seed()
```

```python
# Chargement des données
print("Chargement des données...")
data = load_all_data()
print_dataset_info(data)
```

```python
# Préparation des données
print("Préparation des données...")

# Nettoyer les données target
target_clean = clean_target_data(data['target_train'])

# Préparer le dataset d'entraînement enrichi
df_enriched, imputer = prepare_enriched_dataset(
    data['clinical_train'], 
    data['molecular_train'], 
    target_clean
)

print(f"Dataset enrichi: {df_enriched.shape}")
```

```python
# Visualisation exploratoire
plot_correlation_matrix(df_enriched)
```

```python
# Préparer les features et target pour l'entraînement
X_train, X_test, y_train, y_test, features = prepare_features_and_target(
    df_enriched, target_clean
)

print(f"Features d'entraînement: {X_train.shape}")
print(f"Features de test: {X_test.shape}")
print(f"Nombre de features: {len(features)}")
```

```python
# Entraînement des modèles
print("Entraînement des modèles...")
models = train_and_save_all_models(X_train, y_train)
```

```python
# Évaluation des modèles
print("Évaluation des modèles...")
results, best_model_name = compare_models(models, X_train, y_train, X_test, y_test)

print(f"\nMeilleur modèle: {best_model_name}")
```

```python
# Visualisation des résultats
create_visualization_report(models, results, X_train, features)
```
